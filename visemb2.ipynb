```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the embeddings and sentences
realmente_embeddings = np.load('realmente_embeddings.npy')
realmente_sentences = open('realmente_embeddings_sentences.txt', encoding='utf-8').readlines()

muy_embeddings = np.load('muy_embeddings_2500.npy')
muy_sentences = open('muy_embeddings_sentences_2500.txt', encoding='utf-8').readlines()

rapidamente_embeddings = np.load('rapidamente_embeddings_2500.npy')
rapidamente_sentences = open('rapidamente_embeddings_sentences_2500.txt', encoding='utf-8').readlines()
```

```python
# Concatenate the embeddings and sentences
all_embeddings = np.concatenate([realmente_embeddings, muy_embeddings, rapidamente_embeddings])
all_sentences = realmente_sentences + muy_sentences + rapidamente_sentences

# Normalize the embeddings
all_embeddings = all_embeddings / np.linalg.norm(all_embeddings)

# Calculate pairwise cosine similarities between the embeddings
#similarities = cosine_similarity(all_embeddings)

# Perform k-means clustering
num_clusters = 3 # Define the number of clusters

# Perform PCA on the embeddings
pca = PCA(n_components=2, random_state=0)
embeddings_pca = pca.fit_transform(all_embeddings)

kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(embeddings_pca)
```

```python
# Visualize the clusters in the PCA space
colors = ['purple', 'green', 'blue']

plt.figure(figsize=(15, 15))
scatter = plt.scatter(embeddings_pca[:, 0], embeddings_pca[:, 1], c=[colors[label] for label in kmeans.labels_], alpha=0.5)
plt.title('Clustering of Embeddings')
plt.xlabel('PCA Dimension 1')
plt.ylabel('PCA Dimension 2')
plt.show()
```

```python
realmente_pca = embeddings_pca[:len(realmente_embeddings)]
muy_pca = embeddings_pca[realmente_pca.shape[0]:realmente_pca.shape[0] + muy_embeddings.shape[0]]
rapidamente_pca = embeddings_pca[realmente_embeddings.shape[0] + muy_embeddings.shape[0]:]
```
